### Import libraries

```python
%matplotlib inline
from fastai.basics import *
```

#### Read Data

```python
#define the path
path = Config().data_path()/'mnist'

#open the zip
with gzip.open(path/'mnist.pkl.gz', 'rb') as f:
    ((x_train, y_train), (x_valid, y_valid), _) = pickle.load(f, encoding='latin-1')
    
#visualize an example
plt.imshow(x_train[0].reshape(28,28))

#create tensors for the data
x_train,y_train,x_valid,y_valid = map(torch.tensor, (x_train,y_train,x_valid,y_valid))
n,c = x_train.shape
train_ds = TensorDataset(x_train, y_train)
valid_ds = TensorDataset(x_valid, y_valid)

#create DataBunch
bs=64
data = DataBunch.create(train_ds, valid_ds, bs=bs)
```

#### Define Model Architecture

```python
#Define the model architecture
class MNistNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin1 = nn.Linear(784,200,bias=True)
        self.lin2 = nn.Linear(200,50,bias=True)
        self.lin3 = nn.Linear(50,10,bias=True)
        
    def forward(self,xb):
        x=self.lin1(xb)
        x=F.relu(x)
        x=self.lin2(x)
        x=F.relu(x)
        return self.lin3(x)

#Define Loss function
loss_func = nn.CrossEntropyLoss()
    
#create a Learner
learn = Learner(data, MNistNN(), loss_func=loss_func, metrics=accuracy)

#find learning rate
learn.lr_find()
learn.recorder.plot()
```

#### Train

```python
#Define the update method
def update(model,loss_func,x,y,lr):
    opt = optim.Adam(model.parameters(),lr)
    y_hat = model(x)
    loss = loss_func(y_hat,y)
    loss.backward()
    opt.step()
    opt.zero_grad()
    return loss.item()

#create model
model = MNistNN().cuda()

#train the model
losses = [update(model,loss_func,x,y,lr=1e-3) for x,y in data.train_dl]

#visualize losses
plt.plot(losses);
```

```python
#define learning rate
lr = 1e-3

#fit the model
learn.fit_one_cycle(5,lr)

#visualize learning rate
learn.recorder.plot_lr(show_moms=True)

#visualize losses using fastai library
learn.recorder.plot_losses()
```
